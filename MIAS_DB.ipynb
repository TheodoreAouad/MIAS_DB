```
import numpy as np
import matplotlib.pyplot as plt
```

```
def read_pgm(pgmf):
    """Return a raster of integers from a PGM as an array."""
    pgmf.readline()
    (width, height) = [int(i) for i in pgmf.readline().split()]
    depth = int(pgmf.readline())
    assert depth <= 255

    raster = np.zeros((width,height))
    for j in range(height):
        for i in range(width):
            raster[i,j] = ord(pgmf.read(1))
    return raster
```

```
idx = np.random.randint(1,323)
f = open('mdb' + ('000'+str(idx))[-3:] + '.pgm',"rb")
ras = read_pgm(f)
plt.imshow(ras,"gray")
```
